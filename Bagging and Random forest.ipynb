### Bagging and Random forest

#### Bagging and Random forest using regression

### 1) read housing.csv

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('housing.csv')
df
```

```python

df.isnull().sum()
```

```python
df.columns
```

```python
c=['mainroad','guestroom', 'basement', 'hotwaterheating', 'airconditioning','prefarea', 'furnishingstatus']
for i in c:
    
    from sklearn.preprocessing import LabelEncoder
    le=LabelEncoder()
    le.fit(df[i])
    df[i]=le.transform(df[i])
df
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,[0]]
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train
```

```python
y_train
```

### BaggingRegressor

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor

cart = DecisionTreeRegressor()
model = BaggingRegressor(estimator=cart, n_estimators=100, random_state=7)

model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

RF = RandomForestRegressor(n_estimators=800)

RF.fit(x_train,y_train)
y_pred=RF.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```

### Bagging and Random forest using  classification

### Read pima-indians-diabetes.csv

```python
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
D=pd.read_csv("pima-indians-diabetes.csv",names=names)

D
```

```python
X =D.iloc[:,0:8]
Y =D.iloc[:,[8]]
Y
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size = 0.3, random_state=20)
y_train
```

### Bagging classifier

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
A=DecisionTreeClassifier()
B= BaggingClassifier(estimator=A, n_estimators=100, random_state=7) #Classifier
B.fit(X_train,y_train)
y_pred=B.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

### Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
RF=RandomForestClassifier(n_estimators=800)

RF.fit(X_train,y_train)
y_pred=RF.predict(X_test)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
